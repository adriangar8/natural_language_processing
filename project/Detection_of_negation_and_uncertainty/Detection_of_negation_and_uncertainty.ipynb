*Trigger words method*

```python
# negation trigger words in spanish
negation_cues_spanish = ['no', 'nunca', 'nadie', 'nada', 'jamás', 'tampoco']

# negation trigger words in catalan
negation_cues_catalan = ['no', 'mai', 'ningú', 'res', 'cap', 'tampoc']

# uncertainty trigger words in spanish
uncertainty_cues_spanish = [
    "quizás",
    "tal vez",
    "a lo mejor",
    "probablemente",
    "posiblemente",
    "puede que",
    "puede ser",
    "no estoy seguro",
    "no estoy segura",
    "no tengo claro",
    "no lo sé",
    "no estoy convencido",
    "no estoy convencida",
    "no estoy cierto",
    "no estoy cierta",
    "dudoso",
    "dudosa",
    "incierto",
    "incierta",
    "de repente",
    "en algún momento",
]

# uncertainty trigger words in catalan
uncertainty_cues_catalan = [
    "potser",
    "tal vegada",
    "possiblement",
    "probablement",
    "és possible",
    "no estic segur",
    "no estic segura",
    "no ho tinc clar",
    "no ho sé",
    "no estic convençut",
    "no estic convençuda",
    "no estic cert",
    "no estic certa",
    "dubtós",
    "dubtosa",
    "incert",
    "incerta",
    "de sobte",
    "en algun moment",
]
```

```python
for i, _ in df.iterrows():
  
  # Split the text into words
  text_words = df.at[i, 'text'].split()

  print(text_words)

  # Create an empty list to store the results
  result_list = []

  # Iterate through the text_words list and check if the word is in the words_list
  for index, word in enumerate(text_words):
      if word in (negation_cues_spanish or negation_cues_catalan):
          # If the word is found, create a dictionary entry in the specified format
          entry = {
              'value': {
                  'start': index,
                  'end': index + len(word) - 1,
                  'labels': ["NEG"]
              },
              'id': f'ent{len(result_list)}',
              'from_name': 'label',
              'to_name': 'text',
              'type': 'labels'
          }
          result_list.append(entry)

  result_dict = {'result': result_list}
```
